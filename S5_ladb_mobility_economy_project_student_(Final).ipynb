## 🧩 Explorar, limpiar y preparar los datos

Revisión de cada dataset, su estructura, tipos de datos, columnas, valores faltantes o nulos y realizar las correcciones pertinentes.

```python
# Examinar la estructura de traffic
traffic.info()
traffic.head(3)
```

En la estructura del DF traffic, se observa que:
- Las columnas `UpdateTimeUTC` y `UpdateTimeUTC` se encuentran como tipo texto
- Estas columnas representan fechas y horas, se debe cambiar a datetime
- No hay datos ausentes

```python
# Examinar la estructura de eco
eco.info()
eco.head(3)
```

En la estructura del DF eco, se observa que:
- Las columnas `City GDP/capita`, `Unemployment %`, 'PM2.5 (μg/m³)', 'population (M)', se encuentran como tipo texto(object), las cuales sus resultados son de tipo decimal, por lo tanto se deben cambiar a tipo float
- Tampoco hay datos ausentes

Renombrar columnas

Estandarizar los nombres de columnas para evitar errores y facilitar la unión de los datasets.


```python
# Estandarizar los nombres de las columnas de traffic

traffic = traffic.rename(columns={
    'Country':'country','City':'city','UpdateTimeUTC':'update_time_utc',
    'JamsDelay':'jams_delay','TrafficIndexLive':'traffic_index_live',
    'JamsLengthInKms':'jams_length_kms','JamsCount':'jams_count',
    'TrafficIndexWeekAgo':'traffic_index_week_ago',
    'UpdateTimeUTCWeekAgo':'update_time_utc_week_ago',
    'TravelTimeLivePer10KmsMins':'travel_time_live_per_10kms_mins',
    'TravelTimeHistoricPer10KmsMins':'travel_time_hist_per_10kms_mins',
    'MinsDelay':'mins_delay'})

traffic.columns
```

```python
# Estandarizar los nombres de las columnas de eco

eco = eco.rename(columns={'Year':'year','City':'city','Country':'country',
                          'City GDP/capita':'city_gdp_capita',
                          'Unemployment %':'unemployment_pct',
                          'PM2.5 (μg/m³)':'pm25','Population (M)':'population_m'})

eco.columns
```

### Corregir formatos numéricos y de fecha

**🎯Objetivo:**
Asegurar que las columnas de fechas y valores numéricos estén en formatos correctos para permitir análisis, cálculos y comparaciones precisas.



```python
# Convertir las columnas de traffic a tipo fecha con pd.to_datetime()
traffic['update_time_utc'] = pd.to_datetime(traffic['update_time_utc'],errors="coerce")
traffic['update_time_utc_week_ago'] = pd.to_datetime(traffic['update_time_utc_week_ago'],errors="coerce")

# verificar el cambio
traffic.info()
```

```python
# Limpiar separadores y convertir columnas numéricas en eco

eco['city_gdp_capita']= eco['city_gdp_capita'].astype(str).str.replace('.','').str.replace(',', '.').astype(float)

eco['unemployment_pct'] = eco['unemployment_pct'].astype(str).str.replace('%','').str.replace(',', '.').astype(float)

eco['pm25'] = eco['pm25'].astype(str).str.replace(',', '.').astype(float)

eco['population_m'] = eco['population_m'].astype(str).str.replace(',', '.').astype(float)

# Calcula la población total en unidades absolutas
eco['population'] = eco['population_m']* 1000000

# verificar el cambio
eco.info()

eco.head(3)
```


---

## 🧩 Extraer año y filtrar

Extraer el año permite filtrar la información y trabajar solo con el período más reciente y relevante.

### Extraer columna año y filtrar 2024

**🎯Objetivo**
Identificar el año de cada registro y mantener solo los registros del 2024.


```python
# Extraer el año de las fechas
traffic['year'] = traffic['update_time_utc'].dt.year

# Verificar cambio
traffic.head()
```

```python

# Filtrar los registros del año 2024
traffic_2024 = traffic[traffic["year"]==2024].copy()
eco_2024 = eco[eco["year"]==2024].copy()

# Revisar dataframes nuevos
display(traffic_2024.head())
display(eco_2024.head())


```


---

## 🧩Analizar y resumir datos de movilidad

Como el dataset de tráfico contiene **múltiples registros por ciudad**. Se deberán calcular los promedios anuales por ciudad para simplificar el análisis y obtener una visión más clara de las tendencias generales.

### Calcular promedios de tráfico por ciudad

**🎯Objetivo:**
Obtener una vista consolidada del tráfico promedio por ciudad y año, para analizar patrones generales sin depender de datos diarios.




```python
import pandas as pd

# Calcular los  promedios de trafico por ciudad, país y año
columnas = ["city","country","year"]

traffic_city_year_2024 = traffic_2024.groupby(columnas)[['jams_delay', 'traffic_index_live', 'jams_length_kms', 'jams_count', 'mins_delay', 'travel_time_live_per_10kms_mins', 'travel_time_hist_per_10kms_mins']].agg('mean').reset_index()

traffic_city_year_2024.head()
```

```python
traffic_city_year_2024.sort_values(["jams_delay"], ascending=False)
```

La ciudad con el mayor tiempo promedio de tráfico es México con 2833.05 minutos promedio de retraso


---

## 🧩 Unir movilidad y economía

Combinar datasets para poder analizar cómo se relacionan los indicadores económicos con los de movilidad.



```python
# columnas clave de tráfico y economía
# Agrupar las columnas de cada dataset

left_cols = ['city','country','year','jams_delay','traffic_index_live',
             'jams_length_kms','jams_count','mins_delay',
             'travel_time_live_per_10kms_mins','travel_time_hist_per_10kms_mins']

right_cols = ['city','year','city_gdp_capita','unemployment_pct','pm25','population']

# Uso de .copy() para crear los dos nuevos datasets reducidos
traffic_2024_small = traffic_city_year_2024[left_cols].copy()
eco_2024_small = eco_2024[right_cols].copy()

# Unir datasets
merged = pd.merge(traffic_2024_small, eco_2024_small, on = ['city', 'year'], how = 'inner')

# Mostrar las primeras 5 filas
merged.head(5)
```


---

## 🧩 Visualización y análisis de relaciones

Luego de contar el dataset y unificado, se podrán visualizar patrones por medio de gráficos y entender cómo se relacionan las variables económicas con las de movilidad urbana.



```python
# Crear boxplot para observar el comportamiento de los minutos de congestion JamsDelay

sns.boxplot(data = merged, x = 'jams_delay', showmeans = True)

# obtener promedio para mostrarlo en título
mean_value = merged['jams_delay'].mean()
plt.title(f'Boxplot de JamsDelay (2024)\nPromedio: {mean_value:.2f}')
plt.show()
```

```python
# Crear histograma para ver la distribución de la economía (city_gdp_capita)
# Definir estilo
sns.set_theme(style="ticks")
sns.histplot(eco_2024_small, x="city_gdp_capita", bins=3, color='orange')
plt.title('Distribución de los niveles de PIB per cápita')
plt.xlabel('PIB per cápita (USD)')
plt.ylabel('Frecuencia (Número de ciudades)')
plt.show()

```

```python
# Gráfico de barras para comparar jams_delay y city_gdp_capita por ciudad

fig, ax1 = plt.subplots(figsize=(10, 6))

# Primer eje Y para PIB
ax1.bar(merged['city'], merged['city_gdp_capita'], alpha=0.7, color='green', label='PIB per cápita')
ax1.set_ylabel('PIB per cápita (USD)', color='black')
ax1.tick_params(axis='y', labelcolor='green')

# Segundo eje Y para congestión
ax2 = ax1.twinx()
ax2.plot(merged['city'], merged['jams_delay'], color='orange', marker='o', label='Congestión')
ax2.set_ylabel('Minutos de retraso', color='black')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('PIB per cápita vs Congestión por Ciudad')
plt.tight_layout()
for label in ax1.get_xticklabels():
    label.set_rotation(50)
plt.show()
```


Al comparar el PIB per cápita con los niveles de congestión en las 15 ciudades analizadas, se observa que aproximadamente el 66.6% presenta una combinación de alto PIB y baja congestión, lo que sugiere una posible relación entre menores niveles de congestión y mejores indicadores de productividad.

Sin embargo, esta tendencia no es consistente en todos los casos, ya que existen ciudades que muestran comportamientos contrarios, como bajo PIB con alta congestión, así como alto PIB con altos niveles de congestión.

Esta variabilidad indica que la relación entre ambas variables no es concluyente y que otros factores pueden estar influyendo significativamente en la dinámica urbana. Entre estos factores se pueden considerar el tamaño de la población, el nivel de desarrollo tecnológico, el grado de industrialización, la actividad comercial o la calidad de la infraestructura vial.

Por lo tanto, aunque se identifican ciertas tendencias, es necesario un análisis más profundo para comprender las causas específicas en cada ciudad.

# 🧾 Resumen ejecutivo (plantilla)

**Contexto y objetivo**

Es posible observar que, en algunos casos, menores niveles de congestión podrían estar asociados con un mayor PIB per cápita. Sin embargo, esta relación no es concluyente ni determinante ya que existen múltiples factores adicionales que influyen en el comportamiento de ambas variables, como la infraestructura, la planificación urbana, la densidad poblacional y el nivel de desarrollo económico.
Además, se identifican ciudades donde la productividad es alta y la congestión también lo es, lo que refuerza la idea de que no existe una relación directa y consistente entre estas dos variables.

**Cobertura de datos:**  
  
Se ha filtrado la información por el año 2024, lo cual abarcó 15 ciudades de latinoamérica, donde se inclyuen los siguientes paises: Brasil, Colombia, Argentina, Perú, Méxcio, Uruguay y Chile.

**Metodología (alto nivel):**  

Inicialmente, se realizó una exploración de cada conjunto de datos con el fin de analizar la información contenida y los tipos de datos asociados a cada variable. Esto permitió identificar la necesidad de realizar ajustes, así como definir las columnas de interés para el análisis.
Asimismo, al identificar las columnas en común entre ambos datasets, se procedió a estandarizar sus nombres y renombrarlas, con el objetivo de mejorar la comprensión y facilitar la integración de la información.
  
Para realizar el filtro adecuado, fue necesario crear una nueva columna en el dataset de traffic que contuviera únicamente el año, ya que la variable original incluía información adicional como mes, día y hora. Para el análisis en cuestión, era suficiente trabajar a nivel anual, lo que permitió mayor coherencia al momento de filtrar y combinar los datos.
Posteriormente, se utilizó una unión de tipo inner join para integrar ambos datasets, asegurando que la información estuviera previamente limpia y verificada en cada uno de ellos.
  
Finalmente, para facilitar la interpretación de los datos, se elaboraron distintos gráficos, entre ellos un diagrama de cajas que permite analizar el comportamiento de la congestión en minutos. En este se observa que la mayor concentración de los datos se sitúa aproximadamente entre 100 y 900 minutos de retraso. Asimismo, se identifica un valor extremo que supera los 2500 minutos.
Estos valores atípicos pueden estar influyendo en la distorsión de la media, por lo que su análisis resulta relevante para una interpretación adecuada de los resultados.
Por otro lado, se analiza la distribución de niveles PIB per cápita, a travez de un diagrama de barras, donde se establecen tres columnas y se puede observar mayor concetración de ciudades entre los 10.000 y 18.000 distribucion de PIB (USD)

**Hallazgos iniciales:**  

Se observan casos en los que un mayor PIB per cápita se asocia con menores niveles de congestión, lo cual podría considerarse una situación ideal. Sin embargo, también se presentan escenarios en los que ambas variables son altas o muestran comportamientos inversos. Por lo tanto, y como se mencionó anteriormente, no es adecuado establecer una relación directa entre estas variables, dado que sus valores fluctúan de manera heterogénea.

Ciudades como Ciudad de México, Santiago y Sao Paulo presentan comportamientos atípicos, en los que los niveles de congestión superan lo esperado en relación con sus indicadores económicos. En contraste, Montevideo destaca por registrar un PIB per cápita relativamente alto junto con niveles muy bajos de congestión, evidenciando una dinámica diferente.

**Recomendaciones**  

Ciudades como Ciudad de México podrían tomar como referencia el modelo de movilidad de Montevideo, teniendo en cuenta que presentan niveles de PIB per cápita relativamente similares; sin embargo, Montevideo registra valores significativamente menores de congestión, lo que marca una diferencia importante en términos de eficiencia urbana.
Así mismo, Bogotá también representa un caso relevante de análisis, ya que presenta niveles similares entre congestión y PIB per cápita, lo que sugiere la necesidad de evaluar tanto sus estrategias de movilidad como sus oportunidades de desarrollo económico.
